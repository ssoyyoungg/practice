Gradient 실습

 3x + 1를 미분 하면 => 3

      3 * x^1 =>
      3 * 1 * x^0 =>
      3 * 1 * 1 => 3
 상수는 미분하면 사라짐

```
x = torch.tensor(3.0, requires_grad=True)
y = 2*x**2 + 3*x + 1
y.backward()

print(x.grad)
```

gradient 초기화 안하면 생기는 일

초기화 안하면 앞선 결과 값을 더해버림

초기화 하는 방법 :
x.grad.zero_() 넣기

```
x = torch. tensor(2.0, requires_grad=True)

y1 = x**2
y1.backward()
print("첫번째 : ", x.grad)

y2 = x**3
y2.backward()
print("두번째 : ", x.grad)
```

학습 루프 총집합

```
model = nn.Linear(1, 1)
optimizer =torch.optim.SGD(model.parameters(), lr=0.1)

x = torch.tensor([[1.0], [2.0], [3.0]])
y = torch.tensor([[3.0], [4.0], [5.0]])

for epoch in range(10):
  pred = model(x)
  loss = ((pred-y)**2).mean()

  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

  print(epoch, loss.item())
```

```
print(model.weight.data)
print(model.bias.data)
```
